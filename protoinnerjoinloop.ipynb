```python
import pandas as pd
import os
```

```python
PD_path = '/scratch/jrosen5/applied_proj/sandbox/data/kmerMatrix/PDmatrices'
PRCR_path = '/scratch/jrosen5/applied_proj/sandbox/data/kmerMatrix/PRCRmatrices'

# Initialize an empty DataFrame for our final result
matrix_df = None
```

```python
for directory in [PD_path, PRCR_path]:
    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            # Determine the column name suffix ('NR' or 'R')
            suffix = 'NR' if directory == PD_path else 'R'

            # Extract the base name part from the filename (e.g., 'SRR5088813' from 'SRR5088813_Matrix.csv')
            base_name = os.path.basename(filename).split('_')[0]

            # Construct the full file path
            file_path = os.path.join(directory, filename)

            # Read the CSV file
            df = pd.read_csv(file_path)

            # Rename all columns except 'kmer'
            # Now the columns will be named like 'SRR5088813_NR' or 'SRR5088813_R'
            df.rename(columns={col: f"{base_name}_{suffix}" for col in df.columns if col != 'kmer'}, inplace=True)

            # Merge with the result DataFrame
            if matrix_df is None:
                matrix_df = df
            else:
                matrix_df = pd.merge(matrix_df, df, on='kmer', how='inner')

# Check the final DataFrame
matrix_df.head()
```
